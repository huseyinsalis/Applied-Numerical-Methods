```python
from scipy.integrate import solve_ivp
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
import time

def Chemical(t,Z):
    a=1
    b=0.2
    d=8E-4
    f=2E-8
    C1,C2,C3=Z
    f1=(-a*C1**3)+(b*C3)+(d*C3**4)-(f*(C2**4)*(C1**4))
    f2=d*(C3**4)-(f*(C2**4)*(C1**4))
    f3=a*(C1**3)-b*C3-d*(C3**4)+f*(C2**4)*(C1**4)
    return [f1,f2,f3]

sol = solve_ivp(Chemical, [0, 50], [0.063, 700, 5.6],'RK45',rtol=.01)

plt.figure(1)
plt.plot(sol.t,sol.y[0,:])
plt.legend(['C1: '])

plt.figure(2)
plt.plot(sol.t,sol.y[1,:])
plt.legend(['C2: '])

plt.figure(3)
plt.plot(sol.t,sol.y[2,:])
plt.legend(['C3: '])
```
